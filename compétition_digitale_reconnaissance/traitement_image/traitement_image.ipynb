```python
def image_noir(image):
    for i in range(len(image)):
        for j in range (len(image[0])):
            for k in range(3):
                image[i][j][k]=0
    return Image.fromarray(image)
```

```python
leana_noire=image_noir(tab)
leana_noire.save('leana_noire.png')
```

```python
def image_blanch (image):
    for i in range(len(image)) :
        for j in range (len(image[0])) :
            for k in range(len(image[0][0])):
                image[i][j][k]=255
    return Image.fromarray(image)
```

```python
image_blanch(tab)
```

```python
def image_rouge(image):
    for i in range(len(image)):
        for j in range(len(image[0])):
            for k in range(len(image[0][0])):
                if k==0 :
                    image[i][j][k]=255

                
    return Image.fromarray(image)
```

```python
image_rouge(tab)
```

```python
def image_negative(image):
    for i in range(len(image)):
        for j in range(len(image[0])):
            for k in range(len(image[0][0])):
        
                    image[i][j][k]=255-image[i][j][k]

                
    return Image.fromarray(image)
```

```python
image_negative(tab)
```

```python
def add_pictures(im1,im2):
    s=im1.astype(np.float)+im2.astype(np.float)
    s[s>255]=255
    s[s<0]=0
    return s.astype(np.uint8)
```

```python
plt.imshow(add_pictures(tab,tab))
```

```python
def soustraction_pictures(im1,im2):
    s=im1.astype(np.float)-im2.astype(np.float)
    s[s>255]=255
    s[s<0]=0
    return s.astype(np.uint8)
```

```python
plt.imshow(soustraction_pictures(tab,tab))
```

```python
def plot_histogram(image):
    # Get the 2D array for each of the three colors, and flat it.
    pixels = [p.ravel() for p in np.array_split(image, 3, axis=2)]
    # Produce the histogram for each color and stack them
    style_hist = {'bins': np.arange(-0.5, 256.5, 1.0), 'alpha': 0.5, 'stacked':True}
    plt.hist(pixels, color=['tab:red', 'tab:green', 'tab:blue'], **style_hist)
    return
plot_histogram(tab)
```
